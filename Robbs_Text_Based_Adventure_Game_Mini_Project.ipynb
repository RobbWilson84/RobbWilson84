# What is a Text Based Adventure Game?
---

Early computer games did not have the graphics capabilities or software to create games as we know them now.  This is a good introduction to early computer gaming so they can see how game design and technology has moved on, but it is also an excellent way to develop and consolidate Python 3 skills while having some fun.  The game develops skills for:

* functions (writing, calling and passing in arguments)
* taking input from the user
* print()
* if, elif, else
* equality (==)
* .lower() method to convert user input to lower case
* order of code

# Requirements
____
In this example the following are the functions that are required:

```
start()
load_bear_room()
load_monster_room()
load_diamond_room()
game_over()
win()
```

Each room should start with the intro to the room and provide the choices for the user

The decisions are the choices offered to the user, and depending on the choice they make, the relevant function is loaded.

The order of the functions is important.  For example, if the user chooses option 1 from the load_bear_room(), the game_over() function will be called.  Therefore the game_over() function must be written in the code above the load_bear_room() function, otherwise you will get an error message.

It is good practice to create all your functions in advance and to use pass so the function does not try and run.

```
def start():
  pass
```

# Create Your Adventure
_____

Use the cell below to create your own game as wild as your imagination.

````
def start():
  print("welcome! You find yourself trapped in side of of an abandoned mansion")
  print("It's very dark but you make out three doors in front of you")
  print("Where would you like to go?")
  print("1 for hall")
  print("2 for dungeon")
  print("3 for attic")
  answer = int(input("Where do you want to go? "))

  if answer == 1:
    main_hall()
  elif answer == 2:
    dungeon_deadend()
  elif answer == 3:
    attic_deadend()

def dungeon_deadend():
  print("You find yourself in a very dark room")
  print("You hear strange noises coming from the shadows")
  print("1 for stay")
  print("2 for go")
  stay_go = int(input("Do you stay or go back to the Main Hall? "))

  if stay_go == 1:
      main_hall()
  elif stay_go == 2:
      game_over()

def attic_deadend():
  print("You find yourself in a musty attic room")
  print("a voice starts to speak to you")
  print("1 for listen")
  print("2 to run")
  listen_run = int(input("Do you listen or run? "))

  if listen_run == 1:
    main_hall()
  elif listen_run == 2:
    game_over()

def main_hall():
  print("You are in the main hall")
  print("In front of you is a large table with some items on it")
  print("You can choose to pick up one of the items")
  print("1 for torch")
  print("2 for key")
  item = int(input("Which item do you pick up"))

  if item == 1:
    attic()
  elif item == 2:
    dungeon()

def dungeon():
  print("You are in the dungeon")
  print("In the darkness you notice light coming from a small keyhole")
  print("You begin to hear noise")
  print("1 for key")
  print("2 for run away")
  key_run = int(input("Use your Key or run away? "))

  if key_run == 1:
    win()
  elif key_run == 2:
    game_over()

def attic():
  print("You storm into the attic but forget to check your torch for batteries...")
  print("Shadows descend upon you as the room grows even darker")
  shadow_run = int(input("Do you stay (1) or do you run away (2) "))

  if shadow_run == 1:
    game_over()
  elif shadow_run == 2:
    main_hall()

def game_over():
  print("something from the dark reaches for you...")
  print("...you try to move but are paralysed with fear!")
  print("The last thing you hear is a loud growl as something bites you")
  print("Game Over")
  print("1 = Yes and 2 = No")
  play_again = int(input("Would you like to try again? "))

  if play_again == 1:
    start()
  elif play_again == 2:
    print("Thank you for playing")

def win():
  print("You step into the light...")
  print("Outside of the building you see the sunrise over the hills...")
  print("1 for yes")
  print("2 for no")
  play_again = int(input("Would you like to play again?" ))

  if play_again == 1:
    start()
  elif play_again == 2:
    print("Congratulations...enjoy your life.")

start()
````
